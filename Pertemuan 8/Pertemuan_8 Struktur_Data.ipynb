Representasi Graf dengan Adjacency List

```python
# Membuat graf tak berarah menggunakan adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E'],
}

# Menampilkan graf
def print_graph(graf):
  for node in graf:
    print(f"{node} -> {graf[node]}")

print_graph(graph)
```

Traversal Graf menggunakan BFS

```python
from collections import deque

def bfs(graph, start):
  visited = set()
  queue = deque([start])

  while queue:
    vertex = queue.popleft()
    if vertex not in visited:
      print(vertex, end=" ")
      visited.add(vertex)
      queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])

# Contoh pemanggilan
bfs(graph, 'A')
```

Tarversal Graf menggunakan DFS

```python
def dfs(graph, start, visited=None):
  if visited is None:
    visited = set()

  if start not in visited:
    print(start, end=" ")
    visited.add(start)
    for neighbor in graph[start]:
      dfs(graph, neighbor, visited)

# Contoh pemanggilan
dfs(graph, 'A')
```

Latihan Personal

```python
def count_neighbors(graph):
  result = {}
  for node in graph:
    result[node] = len(graph[node])
  return result

# Contoh penggunaan
graf = {
    'A': ['B', 'c'],
    'B': ['A', 'D'],
    'C': ['A'],
    'D': ['B'],
}
print(count_neighbors(graf))
```

2. Implementasikan BFS dan DFS dari simpul awal pilihan Anda.

```python
from collections import deque

# BFS
def bfs(graph, start):
    visited = set()
    queue = deque([start])

    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            print(vertex, end=" ")
            visited.add(vertex)
            queue.extend(graph[vertex])

# DFS
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    if start not in visited:
        print(start, end=" ")
        visited.add(start)
        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)

# Contoh penggunaan
bfs(graph, 'A')
print()

dfs(graph, 'A')
```

3. Modifikasi fungsi BFS agar mengembalikan list urutan kunjungan bukan hanya mencetak.

```python
from collections import deque

def bfs_list(graph, start):
    visited = set()
    queue = deque([start])
    result = []

    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            visited.add(vertex)
            result.append(vertex)
            queue.extend(graph[vertex])
    return result

# Contoh penggunaan
print(bfs_list(graph, 'A'))
```

4. Buat fungsi find_path(graph, start, end) yang mengembalikan satu jalur dari start ke end jika ada, menggunakan DFS.

```python
def find_path(graph, start, end, path=[]):
    path = path + [start]
    if start == end:
        return path

    for neighbor in graph[start]:
        if neighbor not in path:
            new_path = find_path(graph, neighbor, end, path)
            if new_path:
                return new_path
    return None

# Contoh penggunaan
print(find_path(graph, 'A', 'E'))
```

5. Tambahkan fungsi is_connected(graph) untuk mengecek apakah semua simpul dalam graf saling terhubung (gunakan kombinasi DFS + logika).

```python
def dfs_connected(graph, start, visited):
    visited.add(start)
    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs_connected(graph, neighbor, visited)

def is_connected(graph):
    visited = set()
    start_node = list(graph.keys())[0]
    dfs_connected(graph, start_node, visited)
    return len(visited) == len(graph)

# Contoh penggunaan
print(is_connected(graph))
```
