Binary Tree Manual

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

# Membuat tree secara manual
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

# Fungsi inorder traversal
def inorder(node):
  if node:
    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)

print("Hasil inorder traversal dari binary tree:")
inorder(root)
```

Binary Search Tree

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

class BinarySearchTree:
  def __init__(self):
    self.root = None
  def insert(self, root, data):
    if root is None:
      return Node(data)
    if data < root.data:
      root.left = self.insert(root.left, data)
    else:
      root.right = self.insert(root.right, data)
    return root

bst = BinarySearchTree()
root = None
data_list = [50, 30, 70, 20, 40, 60, 80]
for value in data_list:
  root = bst.insert(root, value)
```

Traversal pada BST

```python
def inorder(node):
  if node:
    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)

def preorder(node):
  if node:
    print(node.data, end=' ')
    preorder(node.left)
    preorder(node.right)

def postorder(node):
  if node:
    postorder(node.left)
    postorder(node.right)
    print(node.data, end=' ')

print("Inorder Traversal:")
inorder(root)
print("\nPreorder Traversal:")
preorder(root)
print("\nPostorder Traversal:")
postorder(root)
```

Pencarian Nilai dalam BST

```python
def search(node, key):
  if node is None or node.data == key:
    return node
  if key < node.data:
    return search(node.left, key)
  return search(node.right, key)

# Uji pencarian
key = 60
result = search(root, key)
if result:
  print(f"{key} ditemukan dalam tree.")
else:
  print(f"{key} tidak ditemukan.")

key = 25
result = search(root, key)
if result:
  print(f"{key} ditemukan dalam tree.")
else:
  print(f"{key} tidak ditemukan.")
```

```python
# Program BST berdasarkan identitas saya

class Node:
  def __init__(self, data):
    self.data = data # menyimpan nilai pada kode
    self.left = None # referensi ke anak kiri
    self.right = None # referensi ke anak kanan

class BinarySearchTree:
  def __init__(self):
    self.root = None # root awal bernilai kosong

  def insert(self, root, data):

    # jika node kosong, dibuat node baru
    if root is None:
      return Node(data)

    # jika data lebih kecil dari root, masuk ke subtree kiri
    if data < root.data:
      root.left = self.insert(root.left, data)

    # jika data lebih besar atau sama, masuk ke subtree kanan
    else:
      root.right = self.insert (root.right, data)
    return root

  def inorder(self, node):
    if node:
      self.inorder(node.left) # menjelajahi subtree kiri
      print(node.data, end=' ') # mencetak node saat ini
      self.inorder(node.right) # menjelajahi subtree kanan

  def preorder(self, node):
    if node:
      print(node.data, end=' ') # mencetak node saat
      self.preorder(node.left) # menjelajahi subtree kiri
      self.preorder(node.right) # menjelajahi subtree kanan

  def postorder(self, node):
    if node:
      self.postorder(node.left) # menjelajahi subtree kiri
      self.postorder(node.right) # menjelajahi subtree kanan
      print(node.data, end=' ') # mencetak node saat ini

  def search(self, node, key):

    # jika node kosong atau data ditemukan
    if node is None or node.data == key:
      return node

    # jika nilai yang dicari lebih kecil, cari di kiri
    if key < node.data:
      return self.search(node.left, key)

    # jika lebih besar, cari di kanan
    return self.search(node.right, key)

bst = BinarySearchTree()
root = None

# penjelasan data:
# 049 = 3 digit terakhir NPM dikonversi menjadi 49 dan diubah menjadi integer karena angka 0 sebagai awalan tidak valid secara sintaks
# 16 = tanggal lahir (DD)
# 02 = bulan lahir (MM) dikonversi sama seperti angka NPM
# ord('E') = nilai ASCII huruf pertama pada nama depan ord() untuk mengubah satu karakter menjadi nilai angka
# ord('u') = nilai ASCII huruf kedua pada nama
# 10 = angka bebas
# 7 = angka bebas
data_list = [int("049"), 16, int("02"), ord('E'), ord('u'), 10, 7]

# memasukkan seluruh data ke dalam BST
for value in data_list:
  root = bst.insert(root, value)

print("Inorder Traversal:")
bst.inorder(root)
print("\nPreorder Traversal:")
bst.preorder(root)
print("\nPostorder Traversal:")
bst.postorder(root)

print("\n")
key = 49
result = bst.search(root, key)
if result:
  print(f"{key} ditemukan dalam tree.")
else:
  print(f"{key} tidak ditemukan.")

key = 67
result = bst.search(root,key)
if result:
  print(f"{key} ditemukan dalam tree.")
else:
  print(f"{key} tidak ditemukan.")
```
