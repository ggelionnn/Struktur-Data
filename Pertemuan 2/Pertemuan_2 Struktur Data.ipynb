1. Buatlah double linked list dengan 3 node didalamnya, kemudian lakukan penyisipan tengah dan hapus dengan kata kunci !

```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None
        self.prevval = None


class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def listprint(self):
        print("Isi Linked List:")
        node = self.head
        while node:
            print(node.dataval)
            node = node.nextval
        print("")

    # Tambah di akhir
    def addinend(self, newdata):
        newnode = Node(newdata)
        if self.head is None:
            self.head = newnode
            return
        last = self.head
        while last.nextval:
            last = last.nextval
        last.nextval = newnode
        newnode.prevval = last

    # Tambah di tengah (setelah node tertentu)
    def addinbetween(self, key, newdata):
        temp = self.head
        while temp:
            if temp.dataval == key:
                break
            temp = temp.nextval
        if temp is None:
            print("Node dengan kata kunci tidak ditemukan")
            return
        newnode = Node(newdata)
        newnode.nextval = temp.nextval
        newnode.prevval = temp
        if temp.nextval:
            temp.nextval.prevval = newnode
        temp.nextval = newnode

    # Hapus berdasarkan kata kunci
    def delete(self, key):
        temp = self.head

        # Jika head yang dihapus
        if temp and temp.dataval == key:
            self.head = temp.nextval
            if self.head:
                self.head.prevval = None
            return
        while temp:
            if temp.dataval == key:
                break
            temp = temp.nextval
        if temp is None:
            print("Node tidak ditemukan")
            return
        if temp.nextval:
            temp.nextval.prevval = temp.prevval
        if temp.prevval:
            temp.prevval.nextval = temp.nextval


# ====== PROGRAM UTAMA ======
dll = DoublyLinkedList()

# 3 node awal
dll.addinend("januari")
dll.addinend("februari")
dll.addinend("maret")

dll.listprint()

# Penyisipan tengah (setelah februari)
dll.addinbetween("februari", "april")
dll.listprint()

# Menghapus dengan kata kunci
dll.delete("februari")
dll.listprint()
```

2. Buatlah circular linked list dengan 3 node didalamnya, kemudian lakukan penyisipan akhir dan hapus akhir !

```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None


class CircularLinkedList:
    def __init__(self):
        self.head = None

    def listprint(self):
        if self.head is None:
            print("List kosong")
            return

        print("Isi Circular Linked List:")
        temp = self.head
        while True:
            print(temp.dataval)
            temp = temp.nextval
            if temp == self.head:
                break
        print("")

    # Tambah di akhir
    def addinend(self, newdata):
        newnode = Node(newdata)
        if self.head is None:
            self.head = newnode
            newnode.nextval = self.head
            return
        temp = self.head
        while temp.nextval != self.head:
            temp = temp.nextval
        temp.nextval = newnode
        newnode.nextval = self.head

    # Hapus di akhir
    def deleteend(self):
        if self.head is None:
            print("List kosong")
            return

        # Jika hanya 1 node
        if self.head.nextval == self.head:
            self.head = None
            return
        temp = self.head
        prev = None
        while temp.nextval != self.head:
            prev = temp
            temp = temp.nextval
        prev.nextval = self.head


# ===== PROGRAM UTAMA =====
cll = CircularLinkedList()

# 3 node awal
cll.addinend("januari")
cll.addinend("februari")
cll.addinend("maret")

cll.listprint()

# Penyisipan akhir
cll.addinend("april")
cll.listprint()

# Hapus akhir
cll.deleteend()
cll.listprint()
```
